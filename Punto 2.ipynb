```python
from collections import defaultdict
import networkx as nx
import numpy
from scipy.cluster import hierarchy
from scipy.spatial import distance
import matplotlib.pyplot as plt


def create_hc(G):
    """Creates hierarchical cluster of graph G from distance matrix"""
    path_length=nx.all_pairs_shortest_path_length(G)
    distances=numpy.zeros((len(G),len(G)))
    for u,p in path_length.items():
        for v,d in p.items():
            distances[u][v]=d
    # Create hierarchical cluster
    Y=distance.squareform(distances)
    Z=hierarchy.complete(Y)  # Creates HC using farthest point linkage
    # This partition selection is arbitrary, for illustrive purposes
    membership=list(hierarchy.fcluster(Z,t=1.15))
    # Create collection of lists for blockmodel
    partition=defaultdict(list)
    for n,p in zip(list(range(len(G))),membership):
        partition[p].append(n)
    return list(partition.values())



# Extract largest connected component into graph H
H=max(nx.connected_component_subgraphs(g_un),key=len)
#print(H.nodes())
for n in H.nodes() :
    H.node[n]['tw_id']=n
        
# Makes life easier to have consecutively labeled integer nodes
H=nx.convert_node_labels_to_integers(H)
# Create parititions with hierarchical clustering
partitions=create_hc(H)
# Build blockmodel graph
BM=nx.blockmodel(H,partitions)


```

```python
#Guardar bloque como atributo
for i in BM.nodes():
    bloque=BM.node[i]['graph']
    for k in bloque.nodes():
        id_original=H.node[k]['tw_id']
        for n in g.nodes():
            g.node[id_original]['bloque']=i
    
```

```python
def seed (G, initial):
    #Agregar el atributo de estado infectado  
    for i in range(len(G.nodes())):
        G.node[G.nodes()[i]]['inf_t']=0
        G.node[G.nodes()[i]]['inf_t+1']=0
    #infectar a los iniciales        
    for i in range(len(initial)):
        G.node[initial[i]]['inf_t']=1
        G.node[initial[i]]['inf_t+1']=1
    return(G)
    
   
            
def SI (G, initial, p):
    #print(initial)
    #N√∫mero de nodos
    nodes= float(len(G.nodes()))
    #inicializar contagio
    seed(G, initial)
    #Garantiza que no se infecte 2 veces en la misma ronda
    infected= set(initial)
    #Lista con la proporci√≥n de nodos infectados
    numinfectedlist=[]
    numinfectedlist.append(float(len(initial)/nodes))
    #inicializa tiempo
    #condicion de parada
    for t in range(1000):
        #actualiza el estado de cada nodo para cada T
        for node in G.nodes():
            G.node[node]['inf_t']=G.node[node]['inf_t+1']
    
#realiza la infecci√≥n en t

        for node in G.nodes():
        #Garantiza que no se infecte en dos rondas
            if G.node[node]['inf_t']<1:
                for neighbors in (G.neighbors(node)):
                    if G.node[neighbors]['inf_t']==1:
                        if rd.random()<p:
                            G.node[node]['inf_t+1']=1
                            infected.add(node)
#Registrar el porcentaje de nodes infectados

        numinfectedlist.append(float(len(infected)/nodes))
        return(numinfectedlist,G)


def seed_gen (G,p):
    random_seq=[]
    #print(len(G.nodes()))
    n=int(round(p*len(G.nodes())))
    #print(n)
    for i in range(n):
        a=rd.randint(0,len(G.nodes())-1)
        random_seq.append(G.nodes()[a])
    return random_seq
```

```python
          
def SIR (G, initial, p, u):
    #print(initial)
    #N√∫mero de nodos
    nodes= float(len(G.nodes()))
    #inicializar contagio
    seed(G, initial)
    #Garantiza que no se infecte 2 veces en la misma ronda
    infected= set(initial)
    recovered=set()
    #Lista con la proporci√≥n de nodos infectados
    numinfectedlist=[]
    numinfectedlist.append(float(len(initial)/nodes))
    #inicializa tiempo
    #condicion de parada
    for t in range(1000):
    #actualiza el estado de cada nodo para cada T
        for node in G.nodes():
            G.node[node]['inf_t']=G.node[node]['inf_t+1']
        for node in G.nodes():
        #Garantiza que no se infecte en dos rondas
            if G.node[node]['inf_t']<1:
                for neighbors in (G.neighbors(node)):
                    if G.node[neighbors]['inf_t']==1:
                    if rd.random()<p:
                        G.node[node]['inf_t+1']=1
                        infected.add(node)
        #Se puede recuperar
            else:
                if rd.random()<u:
                    G.node[node]['inf_t+1']=0
                    recovered=set([node])
                    infected=infected-recovered
    #Registrar el porcentaje de nodes infectados

        numinfectedlist.append(float(len(infected)/nodes))
    return(numinfectedlist,G)
```
