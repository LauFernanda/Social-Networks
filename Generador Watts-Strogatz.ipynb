```python
import networkx as nx
import random as rd
import matplotlib.pyplot as plt

def WS(n,vecinos,p):
    #Genera  grafo
    grafo= nx.Graph()
    k=vecinos
    nodes=n
    #Genera anillo completo
    for v in range(k+1):
        for n in range(nodes):
            if n+v in range(nodes):
                grafo.add_edge(n,n+v)
    for v in range(k):
        for n in range(nodes-k,nodes):
            grafo.add_edge(v,n)
        k=k-1
    #Genera conexiones aleatorias 
    for i,j in grafo.edges():
        L=rd.randrange(nodes)
        if i!=j and i!=L and rd.random()<=p:
            grafo.remove_edge(i,j)
            grafo.add_edge(i,L)
           
    #Genera visualización
    figure=plt.figure(figsize=(20,20))
    pos=nx.shell_layout(grafo)
    nx.draw(grafo, pos=pos, node_size= 300, node_color='green',edge_color='blue')
    plt.show()



    

WS(20,5,0)
```
